# Read an image

Use the function **cv2.imread()** to read an image. The image should be in the working directory or a full path of image should be given.

Second argument is a flag which specifies the way image should be read.
<li> cv2.IMREAD_COLOR  ( 1 )     : If set, always convert image to the 3 channel BGR color image.</li>
<li> cv2.IMREAD_GRAYSCALE  ( 0 ) : If set, always convert image to the single channel grayscale image (codec internal conversion). </li>
<li> cv2.IMREAD_UNCHANGED ( -1 ) : If set, return the loaded image as is (with alpha channel, otherwise it gets cropped). Ignore                                    EXIF orientation. </li>


There are still many option to read images. I provides link here so that you can try by yourself.
<a href="https://docs.opencv.org/3.4/d4/da8/group__imgcodecs.html" > Click here </a>

```python
imagepath = os.path.join(mainpath, 'Cricket team.jpg') # Make exact path 

img = cv2.imread(imagepath,1)                         # Load an color image in grayscale
plt.imshow(img)                                        # Display image through Matplotlib library
```

# Display an image

**cv2.imshow** function used for displaying image in window.

<li> First argument : window name </li>
<li> Second argument : Readable image </li>

Note : You can create as many windows as you wish, but with different window names.

```python
cv2.imshow('Cricket team window',img)            #display image through OpenCV
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

**cv2.waitKey()** is a keyboard joining function to window. Its argument is the time in milliseconds.The function waits for specified milliseconds for any keyboard event then window will automatic closed.
If 0 is passed, it waits indefinitely for a key stroke.
It can also be set to detect specific key strokes like, if key a is pressed etc which we will discuss below.

**cv2.destroyAllWindows()** simply destroys all the windows we created.
If you want to destroy any specific window, use the function **cv2.destroyWindow()** where you pass the exact window name as the argument.

You can specify whether window is resizable or not. It is done with the function **cv2.namedWindow()**. By default, the flag is **cv2.WINDOW_AUTOSIZE**. But if you specify flag to be **cv2.WINDOW_NORMAL**, you can't resize window.

```python
cv2.namedWindow('Demo', cv2.WINDOW_AUTOSIZE)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Write an image

**cv2.imwrite()** function is used for saving an image.

<li> First argument : file name </li>
<li> Second argument : image you want to save </li>

Note: You can save image with any extension. 

```python
cv2.imwrite('cricketdemo.png',img)
```

This will save the image in PNG format in the working directory.
